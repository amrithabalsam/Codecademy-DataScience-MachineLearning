```python
import pandas as pd
```

Import all the files

```python
visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
                   
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

Step 1: Inspect the DataFrames using `print` and `head`

```python
print(visits.head(5))
print(cart.head(5))
print(checkout.head(5))
print(purchase.head(5))
```

Step 2: Left merging visits and cart

```python
v_to_c  = pd.merge(visits,cart,how = 'left')
print(v_to_c.head(5))
```

Step 3: How long is `visits_cart`?

```python
v_to_c['v_to_c_time']  = visits_cart.cart_time - visits_cart.visit_time
print(v_to_c)
```

Step 4: How many timestamps are null for `cart_time`?

```python
v_but_no_c = v_to_c[v_to_c.cart_time.isnull()]
print(v_but_no_c)
v_but_no_c_count  = v_but_no_c.user_id.count()
print(v_but_no_c_count)
```

Step 5: What percentage only visited?

```python
v_to_c['is_cart'] = ~v_to_c.cart_time.isnull()
v_to_c_grouped = v_to_c.groupby('is_cart').user_id.count().reset_index()
print(v_to_c_grouped)


v_to_c_grouped['percent'] =  (v_to_c_grouped.user_id/v_to_c_grouped.user_id.sum())*100
print(v_to_c_grouped)
```

Step 6: What percentage placed a t-shirt in their cart but did not checkout?

```python
cart_to_checkout =  pd.merge(cart,checkout,how='left')
cart_to_checkout['is_checkout']  =  ~cart_to_checkout.checkout_time.isnull()
# print(cart_to_checkout)

c_to_co_grouped = cart_to_checkout.groupby('is_checkout').user_id.count().reset_index()
print(c_to_co_grouped)


c_to_co_grouped['percent'] =  (c_to_co_grouped.user_id/c_to_co_grouped.user_id.sum())*100
print(c_to_co_grouped)
```

Step 7: Merge it all together

```python
all_data  = visits.merge(cart, how='left').merge(checkout, how='left').merge(purchase, how='left')
print(all_data)
```

Step 8: % of users who got to checkout but did not purchase

```python
all_data['is_checkout'] = ~all_data.checkout_time.isnull()
all_data['is_purchase'] = ~all_data.purchase_time.isnull()
# print(all_data)
all_checkouts  = all_data[all_data.is_checkout == True]
# print(all_checkouts)
c_to_p_grouped = all_checkouts.groupby(['is_checkout','is_purchase']).user_id.count().reset_index()
c_to_p_grouped['percent']  = (c_to_p_grouped.user_id/c_to_p_grouped.user_id.sum())*100
print(c_to_p_grouped)
```

Step 9: check each part of the funnel, let's print all 3 of them again

```python
print(v_to_c_grouped)
print(c_to_co_grouped)
print(c_to_p_grouped)
```

*The weakest part of the funnel is clearly getting a person who visited the site to add a tshirt to their cart. Once they've added a t-shirt to their cart it is fairly likely they end up purchasing it. A suggestion could be to make the add-to-cart button more prominent on the front page.*


Step 10: adding new column

```python
all_purchase_data = all_data[all_data.is_purchase == True].reset_index()
all_purchase_data['visit_to_purchase_time']  = all_purchase_data.purchase_time - all_purchase_data.visit_time
print(all_purchase_data.head())
```

Step 11: examine the results

```python
print(all_purchase_data.visit_to_purchase_time)
```

Step 12: average time to purchase

```python
avg_time_to_purchase = all_purchase_data.visit_to_purchase_time.mean()
print(avg_time_to_purchase)
```
